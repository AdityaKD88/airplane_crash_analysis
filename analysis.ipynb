data cleaning

```python
df.Location.replace(np.nan, 'Unknown', inplace= True)
```

```python
df.Operator.replace(np.nan, 'Unknown', inplace= True)
```

```python
df.Route.replace(np.nan, 'Unknown', inplace= True)
```

```python
df.Type.replace(np.nan, 'Unknown', inplace= True)
```

```python
df['Time'] = df['Time'].fillna('00:00')
```

```python
df.dropna(inplace= True)
```

```python
df.isna().sum()
```

```python
df['Time'] = df['Time'].str.replace('c: ', '')
df['Time'] = df['Time'].str.replace('c:', '')
df['Time'] = df['Time'].str.replace('c', '')
df['Time'] = df['Time'].str.replace('12\'20', '12:20')
df['Time'] = df['Time'].str.replace('18.40', '18:40')
df['Time'] = df['Time'].str.replace('0943', '09:43')
df['Time'] = df['Time'].str.replace('22\'08', '22:08')
df['Time'] = df['Time'].str.replace('114:20', '00:00')
```

```python
df['Ground'] = df['Aboard'] - df['Fatalities']
```

```python
df['DateTime'] = df['Date'] + ' ' + df['Time']
```

```python
df['DateTime'] = pd.to_datetime(df['DateTime'])
```

visualization

```python
#Distribution of fatalities by year
plt.figure(figsize= (18, 8))
years = df['DateTime'].dt.year
_= plt.plot(df['Fatalities'], years, 'o')
_= plt.xlabel('Fatalities')
_= plt.ylabel('Years')
plt.show()
```

```python
# #Distribution of no of accidents by year
Temp = df.groupby(df.DateTime.dt.year)[['Date']].count()
Temp = Temp.rename(columns={"Date": "Count"})

plt.figure(figsize=(12,6))
plt.style.use('bmh')
plt.plot(Temp.index, 'Count', data=Temp, color='m', marker = "o", linewidth=1)
plt.xlabel('Year', fontsize=10)
plt.ylabel('Count', fontsize=10)
plt.title('Yearly Accidents', loc='Center', fontsize=18)
plt.show()
```

```python
#Distribution of fatalities by month
month_count = df.groupby(df.DateTime.dt.month)[['Date']].count()

plt.figure(figsize= (16, 8))
plt.plot(range(len(month_count)), month_count['Date'], marker= '.', color= 'orange')
plt.xticks(range(len(month_count)), ['Jan', 'Feb', 'March', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.xlabel('Month')
plt.ylabel('Count of Accidents')
plt.show()
```

```python
#Distribution of fatalities by day of month
day_count = df.groupby(df.DateTime.dt.day)[['Date']].count()

plt.figure(figsize= (16, 8))
sns.barplot(day_count.index, day_count['Date'], color ='lightskyblue')
plt.xlabel('Day')
plt.ylabel('Total Accidents');
```

```python
df['Day'] = df.DateTime.dt.day_name()
day = df.groupby('Day')[['Fatalities', 'Aboard']].sum()
day['Survival Ratio'] = (day['Aboard'] - day['Fatalities']) / day['Aboard']
day = day.sort_index()
```

```python
_,  ax = plt.subplots(figsize=(17,10))
ax.plot(day.index, day['Fatalities'], marker= '.')
ax2 = ax.twinx()
ax2.plot(day.index, day['Survival Ratio'], marker= '.', linestyle= '--', color= 'red')

plt.xlabel('Day of Week')
ax.set_xticklabels(['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'])
ax.set_ylabel('Total Fatalities')
ax2.set_ylabel('Survival Ratio')
plt.show()
```

```python
#Distribution of accidents by sector
sector = df.groupby('Sector')[['Fatalities']].sum().reset_index()
plt.figure(figsize= (16, 10))
plt.pie(sector['Fatalities'], labels= sector['Sector'],radius= 1.2, shadow= True, autopct= '%.2f');
```

```python
#Poisson distribution of days passed between accidents
ndf = df.sort_values('DateTime').reset_index()
```

```python
timedelta = []
n= 0
for i in range(len(ndf)-1):
    timedelta.append(ndf['DateTime'][n+1] - ndf['DateTime'][n])
    n+= 1
```

```python
days = []
for i in timedelta :
	days.append(i.days)
days.append(0)
ndf['Time Since'] = days 
```

```python
plt.figure(figsize= (16, 8))
plt.xlim(0, 85)
ndf['Time Since'].hist(bins= 1600);
```

```python
def  ecdf(data) :
    n = len(data)
    x = np.sort(data)
    y = np.arange(1, n+1) / n
    
    return x, y
```

```python
x, y = ecdf(ndf['Time Since'])

plt.figure(figsize= (16, 8))
plt.xlim(0, 200)
_= plt.plot(x, y, marker='.', linestyle= 'none')
plt.margins(y= 0.03)
_= plt.xlabel('Days Passed')
_= plt.ylabel('ECDF')
plt.show()
```

```python
Fatalities = df.groupby(df.DateTime.dt.year).sum()
Fatalities['Proportion'] = Fatalities['Fatalities'] / Fatalities['Aboard']

plt.figure(figsize=(15,6))
plt.subplot(1, 2, 1)
plt.fill_between(Fatalities.index, 'Aboard', data=Fatalities, color="green", alpha=0.2)
plt.plot(Fatalities.index, 'Aboard', data=Fatalities, marker = ".", color="red", alpha=0.6, linewidth=1)
plt.fill_between(Fatalities.index, 'Fatalities', data=Fatalities, color="magenta", alpha=0.2)
plt.plot(Fatalities.index, 'Fatalities', data=Fatalities, color="blue", marker = ".", alpha=0.6, linewidth=1)
plt.legend(fontsize=10)
plt.xlabel('Year', fontsize=10)
plt.ylabel('Amount of people', fontsize=10)
plt.title('Yearly Human Deaths', loc='Center', fontsize=14)

plt.subplot(1, 2, 2)
plt.plot(Fatalities.index, 'Proportion', data=Fatalities, marker = ".", color = 'red', linewidth=1)
plt.xlabel('Year', fontsize=10)
plt.ylabel('Ratio', fontsize=10)
plt.title('Fatalities Ratio by Year', loc='Center', fontsize=14)
plt.tight_layout()
plt.show()
```
